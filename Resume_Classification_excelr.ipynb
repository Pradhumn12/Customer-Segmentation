```python
import pandas as pd
import numpy as np
```

```python
resume_data = pd.read_excel('/content/feedbacks_with_gemini (1).xlsx')
```

```python
resume_data
```

```python
cleaned_data =resume_data.drop(['file_name','text','original_word_count','cleaned_word_count','gemini_summary','gemini_experience_years','gemini_rating','gemini_feedback'],axis=1)
```

```python
print(cleaned_data)
```

```python
cleaned_data.isnull()
```

```python
cleaned_data.dropna(inplace=True)
```

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText)
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
cleaned_data['cleaned_text'] = cleaned_data['cleaned_text'].apply(lambda x: cleanResume(x))
```

```python
cleaned_data
```

```python
## Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le = LabelEncoder()
cleaned_data['LabelEncoding'] = Le.fit_transform(cleaned_data['category'])
cleaned_data
```

```python
cleaned_data.describe()
```

```python
cleaned_data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
x = cleaned_data['cleaned_text'].values
y = cleaned_data['category'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
tfidf_vector = TfidfVectorizer(sublinear_tf=True,stop_words='english')
x_train_tfidf = tfidf_vector.fit_transform(x_train)
```

```python
x_train_tfidf.shape
```

```python
### Support Vector Machine
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import f1_score, classification_report, precision_score, recall_score
```

```python
model_svm = SVC()
model_svm.fit(x_train_tfidf,y_train)
```

```python
prediction = model_svm.predict(x_train_tfidf)
prediction
```

```python
accuracy_svm = accuracy_score(y_train, prediction)
print('Accuracy of training set : {:.2f}'.format(model_svm.score(x_train_tfidf, y_train)))
# Classification Report
print("Classification Report:")
print(classification_report(y_train, prediction))
```

```python
## Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators=200)
model_rf.fit(x_train_tfidf, y_train)
```

```python
prediction_rf = model_rf.predict(x_train_tfidf)
prediction_rf
```

```python
accuracy_rf = accuracy_score(y_train,prediction_rf)
print('Accuracy of training set : {:.2f}'.format(model_rf.score(x_train_tfidf, y_train)))
# Classification Matrix
print("Classification Report:")
print(classification_report(y_train, prediction_rf))
```
